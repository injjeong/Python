```python
import tkinter as tk
from tkinter import messagebox as msg
```

```python
#베이스 만들기
rt=tk.Tk()
rt.title('Calcuator')

#크기 고정 시키기 위한
rt.geometry("300x170")  
rt.resizable(False,False) #창 안움직임

#입력 및 결과 표시 창
lab = tk.Label(rt, text="입력")
lab.grid(row=0, column=0)

dis = tk.Entry(rt, width=33, justify="right") #숫자가 오른족에 나타나도록
dis.grid(row=0, column=1, columnspan=5)

lab2 = tk.Label(rt, text="결과")
lab2.grid(row=1, column=0)

dis2 = tk.Entry(rt, width=33, justify="right")
dis2.grid(row=1, column=1, columnspan=5)

temp = 0

def cal():
 
    global temp
    value = tk.Entry.get(dis)
    if (value==''):
        tk.messagebox.showwarning("error", "숫자를 입력하시오")
    else:    
        res = eval(value)
        print(res)

        if not(value==''):
            temp = temp + res
            
            print(temp, "temp")
            #print(res, "res")
            dis.delete(0, 'end')   #문자열 없애고
            dis2.delete(0, 'end')   
            dis2.insert(0, temp)   #결과 보여주기
             
        elif(value == 0):                
            temp = temp + value
            
            print(temp, "sum")
            dis.delete(0, 'end')   #문자열을 없애고
            dis2.delete(0, 'end')
            dis2.insert(0, res)    #결과를 보여주기
           
        
#숫자와 연산자 버튼 처리함수
def btn_prs(value):
    dis.insert("end", value) #end는 오른쪽, 입력 창으로 숫자 전송
    print(value, "pressed")
    
#부호 버튼 처리함수
def sign_prs(sign):
    global temp
    if sign == 'CE':  #한 숫자만 지움
        t = tk.Entry.get(dis)
        x = ''
        for i in range(len(t)-1):
            x+=t[i]
            
        dis.delete(0, 'end')
        dis.insert(0, x)
        print(sign, "pressed")
        
    elif sign == 'C': #전체 다 지움
        dis.delete(0, 'end')
        dis2.delete(0, 'end')
        temp = 0
        
        print(sign, "pressed")
        
    elif sign == '+/-':
        a = tk.Entry.get(dis)
        b = int(a) * -1
        dis.delete(0, 'end')
        dis.insert(0, b)
        print(sign, "pressed")
        
    elif sign == '1/x':
        c = tk.Entry.get(dis)
        if (int(c)==0):
            tk.messagebox.showwarning("error", "0으로 나눌 수 없습니다")
        else:
            d = 1/int(c)
            dis.delete(0, 'end')
            dis.insert(0, d)
        print(sign, "pressed")
        
#결과창 닫기
#def quit():
#    rt.destroy()
#    rt.quit()

# <숫자 버튼>

#첫번째
bnt7 = tk.Button(rt, text="7", width=7, command = lambda:btn_prs('7'))
bnt7.grid(row=2, column=0)

bnt8 = tk.Button(rt, text="8", width=7, command = lambda:btn_prs('8'))
bnt8.grid(row=2, column=1)

bnt9 = tk.Button(rt, text="9", width=7, command = lambda:btn_prs('9'))
bnt9.grid(row=2, column=2)

bnt_ce = tk.Button(rt, text="CE", width=7, command = lambda:sign_prs('CE'))
bnt_ce.grid(row=2, column=3)
 
bnt_c = tk.Button(rt, text="C", width=7, command = lambda:sign_prs('C'))
bnt_c.grid(row=2, column=4)


#두번째
bnt4 = tk.Button(rt, text="4", width=7, command = lambda:btn_prs('4'))
bnt4.grid(row=3, column=0)

bnt5 = tk.Button(rt, text="5", width=7, command = lambda:btn_prs('5'))
bnt5.grid(row=3, column=1)

bnt6 = tk.Button(rt, text="6", width=7, command = lambda:btn_prs('6'))
bnt6.grid(row=3, column=2)

bnt_mul = tk.Button(rt, text="*", width=7, command = lambda:btn_prs('*'))
bnt_mul.grid(row=3, column=3)

bnt_div = tk.Button(rt, text="/", width=7, command = lambda:btn_prs('/'))
bnt_div.grid(row=3, column=4)
 
#세번째
bnt1 = tk.Button(rt, text="1", width=7, command = lambda:btn_prs('1'))
bnt1.grid(row=4, column=0)

bnt2 = tk.Button(rt, text="2", width=7, command = lambda:btn_prs('2'))
bnt2.grid(row=4, column=1)

bnt3 = tk.Button(rt, text="3", width=7, command = lambda:btn_prs('3'))
bnt3.grid(row=4, column=2)

bnt_add = tk.Button(rt, text="+", width=7, command = lambda:btn_prs('+'))
bnt_add.grid(row=4, column=3)

bnt_sub = tk.Button(rt, text="-", width=7, command = lambda:btn_prs('-'))
bnt_sub.grid(row=4, column=4)

#네번째
bnt_ps = tk.Button(rt, text="1/x", width=7, command = lambda:sign_prs('1/x'))
bnt_ps.grid(row=5, column=0)

bnt0 = tk.Button(rt, text="0", width=7, command = lambda:btn_prs('0'))
bnt0.grid(row=5, column=1)

bnt_dot = tk.Button(rt, text=".", width=7, command = lambda:btn_prs('.'))
bnt_dot.grid(row=5, column=2)

bnt_r = tk.Button(rt, text="=", width=7, command = cal)
bnt_r.grid(row=5, column=3)
bnt_r['fg'] = 'red'

bnt_b = tk.Button(rt, text="+/-", width=7, command = lambda:sign_prs('+/-'))
bnt_b.grid(row=5, column=4)
   
rt.mainloop()
```
